```python
# Import necessary libraries
import os
import time
import copy
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm

import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader

import torchvision
from torchvision import datasets, models, transforms
from torchvision.transforms import v2

from PIL import Image
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Set device
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Data transforms - ViT typically uses 224x224 input size
data_transforms = {
    'train': transforms.Compose([
        transforms.RandomResizedCrop(224),
        transforms.RandomHorizontalFlip(),
        transforms.RandomRotation(15),
        transforms.ColorJitter(brightness=0.1, contrast=0.1, saturation=0.1),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ]),
    'val': transforms.Compose([
        transforms.Resize(256),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ]),
}

# Load the dataset
# Assuming same data path as in the ensemble code
data_dir = 'C:/Users/merye/Desktop/IP102_Wheat_Cereal_Pests'
image_datasets = {x: datasets.ImageFolder(os.path.join(data_dir, x), data_transforms[x]) 
                  for x in ['train', 'val']}
dataloaders = {x: DataLoader(image_datasets[x], batch_size=32, shuffle=True, num_workers=4)
              for x in ['train', 'val']}
dataset_sizes = {x: len(image_datasets[x]) for x in ['train', 'val']}
class_names = image_datasets['train'].classes
num_classes = len(class_names)

print(f"Dataset loaded with {len(class_names)} classes: {class_names}")
print(f"Training set: {dataset_sizes['train']} images")
print(f"Validation set: {dataset_sizes['val']} images")

# Visualize a few training images
def imshow(inp, title=None):
    """Display image from tensor."""
    inp = inp.numpy().transpose((1, 2, 0))
    mean = np.array([0.485, 0.456, 0.406])
    std = np.array([0.229, 0.224, 0.225])
    inp = std * inp + mean
    inp = np.clip(inp, 0, 1)
    plt.imshow(inp)
    if title is not None:
        plt.title(title)
    plt.pause(0.001)

# Get a batch of training data
inputs, classes = next(iter(dataloaders['train']))

# Make a grid from batch
out = torchvision.utils.make_grid(inputs[:16])
imshow(out, title=[class_names[x] for x in classes[:16]])

# Load pre-trained Vision Transformer model
print("Loading pre-trained ViT model...")
vit_model = torchvision.models.vit_b_16(weights='IMAGENET1K_V1')

# Modify the head to match our number of classes
vit_model.heads = nn.Sequential(
    nn.Linear(vit_model.hidden_dim, 512),
    nn.LayerNorm(512),
    nn.Linear(512, num_classes)
)

# Move model to device
vit_model = vit_model.to(device)

# Define the training and evaluation functions
def train_model(model, criterion, optimizer, scheduler, num_epochs=20):
    """Train the model and return the best model weights."""
    since = time.time()
    
    best_model_wts = copy.deepcopy(model.state_dict())
    best_acc = 0.0
    
    history = {'train_loss': [], 'val_loss': [], 'train_acc': [], 'val_acc': []}
    
    for epoch in range(num_epochs):
        print(f'Epoch {epoch+1}/{num_epochs}')
        print('-' * 10)
        
        # Each epoch has a training and validation phase
        for phase in ['train', 'val']:
            if phase == 'train':
                model.train()  # Set model to training mode
            else:
                model.eval()   # Set model to evaluate mode
                
            running_loss = 0.0
            running_corrects = 0
            
            # Iterate over data
            for inputs, labels in tqdm(dataloaders[phase], desc=phase):
                inputs = inputs.to(device)
                labels = labels.to(device)
                
                # Zero the parameter gradients
                optimizer.zero_grad()
                
                # Forward pass
                with torch.set_grad_enabled(phase == 'train'):
                    outputs = model(inputs)
                    _, preds = torch.max(outputs, 1)
                    loss = criterion(outputs, labels)
                    
                    # Backward pass + optimize only in training phase
                    if phase == 'train':
                        loss.backward()
                        optimizer.step()
                        
                # Statistics
                running_loss += loss.item() * inputs.size(0)
                running_corrects += torch.sum(preds == labels.data)
                
            epoch_loss = running_loss / dataset_sizes[phase]
            epoch_acc = running_corrects.double() / dataset_sizes[phase]
            
            # Record history
            if phase == 'train':
                history['train_loss'].append(epoch_loss)
                history['train_acc'].append(epoch_acc.item())
                if scheduler:
                    scheduler.step()
            else:
                history['val_loss'].append(epoch_loss)
                history['val_acc'].append(epoch_acc.item())
            
            print(f'{phase} Loss: {epoch_loss:.4f} Acc: {epoch_acc:.4f}')
            
            # Deep copy the model if it's the best
            if phase == 'val' and epoch_acc > best_acc:
                best_acc = epoch_acc
                best_model_wts = copy.deepcopy(model.state_dict())
                
        print()
    
    time_elapsed = time.time() - since
    print(f'Training complete in {time_elapsed // 60:.0f}m {time_elapsed % 60:.0f}s')
    print(f'Best val Acc: {best_acc:.4f}')
    
    # Plot training history
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(history['train_loss'], label='Train')
    plt.plot(history['val_loss'], label='Validation')
    plt.title('Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    
    plt.subplot(1, 2, 2)
    plt.plot(history['train_acc'], label='Train')
    plt.plot(history['val_acc'], label='Validation')
    plt.title('Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.show()
    
    # Load best model weights
    model.load_state_dict(best_model_wts)
    return model, history

def evaluate_model(model, dataloader):
    """Evaluate model performance on provided dataloader."""
    model.eval()
    all_preds = []
    all_labels = []
    
    with torch.no_grad():
        for inputs, labels in tqdm(dataloader, desc="Evaluating"):
            inputs = inputs.to(device)
            labels = labels.to(device)
            
            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)
            
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
    
    # Calculate metrics
    accuracy = accuracy_score(all_labels, all_preds)
    print(f'Model Accuracy: {accuracy*100:.2f}%')
    
    # Print classification report
    print("\nClassification Report:")
    print(classification_report(all_labels, all_preds, target_names=class_names))
    
    # Plot confusion matrix
    cm = confusion_matrix(all_labels, all_preds)
    plt.figure(figsize=(12, 10))
    plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
    plt.title('Confusion Matrix')
    plt.colorbar()
    tick_marks = np.arange(len(class_names))
    plt.xticks(tick_marks, class_names, rotation=45)
    plt.yticks(tick_marks, class_names)
    
    # Label the confusion matrix
    fmt = 'd'
    thresh = cm.max() / 2.
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            plt.text(j, i, format(cm[i, j], fmt),
                     ha="center", va="center",
                     color="white" if cm[i, j] > thresh else "black")
    
    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.show()
    
    return accuracy

# Set up training parameters for ViT
# First, freeze most of the pretrained model and only train the head
for name, param in vit_model.named_parameters():
    if "heads" not in name:  # Freeze all layers except heads
        param.requires_grad = False
        
# Define loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.AdamW(vit_model.parameters(), lr=0.001, weight_decay=0.01)
scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=5, gamma=0.1)

# Train the model (first stage - just the head)
print("First stage training - only training the classifier head")
vit_model, history_stage1 = train_model(vit_model, criterion, optimizer, scheduler, num_epochs=10)

# Second stage - fine-tune the entire model
print("Second stage training - fine-tuning the entire model")
for param in vit_model.parameters():
    param.requires_grad = True

# Use a lower learning rate for fine-tuning
optimizer = optim.AdamW(vit_model.parameters(), lr=0.0001, weight_decay=0.01)
scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=10)

vit_model, history_stage2 = train_model(vit_model, criterion, optimizer, scheduler, num_epochs=10)

# Save the model
torch.save({
    'model_state_dict': vit_model.state_dict(),
    'optimizer_state_dict': optimizer.state_dict(),
    'class_names': class_names,
    'num_classes': num_classes
}, 'vit_insect_classifier.pth')

print("Model saved to 'vit_insect_classifier.pth'")

# Evaluate the model
print("Evaluating model on validation set...")
val_accuracy = evaluate_model(vit_model, dataloaders['val'])

# Visualize some predictions
def visualize_model_predictions(model, num_images=6):
    """Display model predictions for a few images."""
    was_training = model.training
    model.eval()
    images_so_far = 0
    fig = plt.figure(figsize=(12, 8))
    
    with torch.no_grad():
        for i, (inputs, labels) in enumerate(dataloaders['val']):
            inputs = inputs.to(device)
            labels = labels.to(device)
            
            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)
            
            for j in range(inputs.size()[0]):
                images_so_far += 1
                ax = plt.subplot(num_images//3 + 1, 3, images_so_far)
                ax.axis('off')
                ax.set_title(f'Predicted: {class_names[preds[j]]}\nActual: {class_names[labels[j]]}',
                           color=("green" if preds[j] == labels[j] else "red"))
                imshow(inputs.cpu().data[j])
                
                if images_so_far == num_images:
                    model.train(mode=was_training)
                    return
        model.train(mode=was_training)

# Visualize model predictions
visualize_model_predictions(vit_model)

# Compare with pre-trained models used in your ensemble
print("\nPerformance Comparison:")
print(f"Vision Transformer: {val_accuracy*100:.2f}%")

# Function to make predictions on new images
def predict_image(model, image_path):
    """Make prediction on a single image."""
    img = Image.open(image_path)
    img_transform = data_transforms['val']
    img_tensor = img_transform(img).unsqueeze(0).to(device)
    
    model.eval()
    with torch.no_grad():
        output = model(img_tensor)
        probabilities = F.softmax(output, dim=1)[0]
        _, predicted_idx = torch.max(output, 1)
        
    predicted_class = class_names[predicted_idx.item()]
    confidence = probabilities[predicted_idx].item() * 100
    
    # Get top 3 predictions
    top3_prob, top3_idx = torch.topk(probabilities, 3)
    top3_classes = [class_names[idx.item()] for idx in top3_idx]
    top3_probs = [prob.item() * 100 for prob in top3_prob]
    
    # Display image with predictions
    plt.figure(figsize=(8, 10))
    plt.subplot(2, 1, 1)
    img = transforms.Resize((224, 224))(img)
    plt.imshow(np.array(img))
    plt.title(f"Predicted: {predicted_class} ({confidence:.2f}%)")
    plt.axis('off')
    
    # Display top 3 predictions as bar chart
    plt.subplot(2, 1, 2)
    plt.barh(top3_classes, top3_probs, color='skyblue')
    plt.xlabel('Confidence (%)')
    plt.title('Top 3 Predictions')
    plt.xlim(0, 100)
    plt.tight_layout()
    plt.show()
    
    return predicted_class, confidence, zip(top3_classes, top3_probs)

# Usage example (uncomment to use):
# predict_image(vit_model, "path_to_test_image.jpg")
```
